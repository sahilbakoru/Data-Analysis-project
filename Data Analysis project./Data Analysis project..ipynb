```python
data=pd.read_csv("countries.csv")
```

```python
data
```

```python
us= data[data.country=='United States']
```

```python
India= data[data.country=='India']
```

```python
china= data[data.country=='China']
```

```python
plt.plot(us.year, us.population/10**6)
plt.plot(India.year, India.population/10**6)
plt.plot(china.year, china.population/10**6)
plt.legend(['us', 'India','china'])
plt.xlabel('year')
plt.ylabel("population (in millons)")
plt.show()
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.plot(India.year, India.population / India.population.iloc[0] * 100)
plt.legend(['United States', 'China', 'India'])
plt.xlabel('year')
plt.ylabel('population growth in % (first year = 100)')
plt.show()
```
